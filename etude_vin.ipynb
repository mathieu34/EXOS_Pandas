```python
col = vini2.apply ( lambda row : ((row["production commercialisable : Volumes AOP, IGP, VSIG et vins aptes (hl)"] *100) / row["total (hl)"]), axis = 1 )
vini2["production commercialisable en %"] = col  
vini2
```

```python
vini2["A.O.P (hl)"] = vini2.iloc[:,[6,7,8,9]].sum(axis=1)
vini2["I.G.P (hl)"] =  vini2.iloc[:,[10,11,12,13]].sum(axis=1)
vini2["V.S.I.G (hl)"] =  vini2.iloc[:,[14,15,16]].sum(axis=1)

col1 = vini2.apply ( lambda row : ((row["A.O.P (hl)"] *100) / (row["production commercialisable : Volumes AOP, IGP, VSIG et vins aptes (hl)"] - row["quantites de vins aptes a la production de Cognac ou d'Armagnac (hl)"])), axis = 1 )
vini2["A.O.P en %"] = col1  
col2 = vini2.apply ( lambda row : ((row["I.G.P (hl)"] *100) / (row["production commercialisable : Volumes AOP, IGP, VSIG et vins aptes (hl)"] - row["quantites de vins aptes a la production de Cognac ou d'Armagnac (hl)"])), axis = 1 )
vini2["I.G.P en %"] = col2  
col3 = vini2.apply ( lambda row : ((row["V.S.I.G (hl)"] *100) / (row["production commercialisable : Volumes AOP, IGP, VSIG et vins aptes (hl)"] - row["quantites de vins aptes a la production de Cognac ou d'Armagnac (hl)"])), axis = 1 )
vini2["V.S.I.G en %"] = col3  

vini2
```
